# 장르

```python
request = requests.get('https://www.webtoons.com/en/genre')
html = request.text

soup = BeautifulSoup(html, 'html.parser')
genre = soup.find_all('ul', 'snb _genre' )
for gen in genre:    
    print(gen.text)
    genre_column=gen.text.split('\n')
genre_column.remove('')
genre_column.remove('')
genre_column
```

```python
genre_column
```

```python
df = pd.DataFrame(columns = ['title','author','grade','genre'])
df
```

```python
df_title,df_author, df_grade = [],[],[]

for i in title:
    df_title.append(i.text)
for j in author:
    df_author.append(j.text)
for k in grade:
    df_grade.append(k.text)
df['title'] = df_title
df['author'] = df_author
df['grade'] = df_grade
df
```

```python
print(df[df['title'].str.contains('The Devil is')], '드라마')
print(df[df['title'].str.contains('CARL')], '판타지')
print(df[df['title'].str.contains('Backchannel')], '코미디')
print(df[df['title'].str.contains('CARL')], '액션')
print(df[df['title'].str.contains('#Blessed')], '슬라이스오브라이프')
print(df[df['title'].str.contains('CARL')], '로맨스')
print(df[df['title'].str.contains('Backchannel')] , '슈퍼히어로')
print(df[df['title'].str.contains('Room of Swords')] , 'Sci-Fi')
print(df[df['title'].str.contains('Uriah')] , 'Thriller')
print(df[df['title'].str.contains('Lali')] , 'Supernatural')
print(df[df['title'].str.contains('Third Shift Society')] , 'Mystery')
```

# 장르 원핫인코딩

```python
# print(genre_column)
genre_col = []
for i in genre_column:
    genre_col.append(i[0]+i[1:].lower())
print(genre_col)
print(list(df_data['genre'].unique()))
```

```python
df_Only_one = df_data.iloc[:, :3]
df_Only_one = df_Only_one.drop_duplicates()
for i in genre_col:
    df_Only_one[i] = 0
df_Only_one
```

```python
# df_Only_one[df_Only_one['title']==i].index
# df_Only_one.loc[df_Only_one[df_Only_one['title']==i].index, j] = 1
```

```python
# 각 웹툰 장르 원핫 인코딩
for i in df_Only_one['title'].unique():
    genres = list(df_data[df_data['title']== i]['genre'])
    for j in genres:    
        if j in df_Only_one.columns:
            df_Only_one.loc[df_Only_one[df_Only_one['title']==i].index, j] = 1
        else:
            df_Only_one.loc[df_Only_one[df_Only_one['title']==i].index, 'Others'] = 1
```

```python
df_Only_one = df_Only_one.reset_index(drop=True)
df_Only_one
```

```python
# 각 장르의 작품 수 (작품 중복 허락)
count, genre = [],[]
for i in df_Only_one.columns[3:]:
#     print(df_Only_one[i].sum())
    genre.append(i)
    count.append(df_Only_one[i].sum())
```

```python
plt.figure(figsize=(16,9))
plt.bar(genre, height=count)
```
