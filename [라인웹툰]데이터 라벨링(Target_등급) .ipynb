# 필요 라이브러리

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# 웹크롤링(스크래핑) 함수 

```python

def scrap_html(url):
    request = requests.get(url)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')
    return soup


url = 'https://www.webtoons.com/en/dailySchedule'
```

# 요일별 인기순 웹툰 스크래핑

## 월요일 웹툰 

```python
MON = scrap_html(url).find('div', {
    'class': 'daily_section _list_MONDAY'
}).find_all('p', "subj")
print(len(MON))
MON_WT = []
for i in MON:
    MON_WT.append(i.text)
```

## 화요일 웹툰 

```python
TUE = scrap_html(url).find('div', {
    'class': 'daily_section _list_TUESDAY'}).find_all('p', "subj")
print(len(TUE))
TUE_WT = []
for i in TUE:
    TUE_WT.append(i.text)
```

## 수요일 웹툰 

```python
WED = scrap_html(url).find('div', {
    'class': 'daily_section _list_WEDNESDAY'
}).find_all('p', "subj")
print(len(WED))
WED_WT = []
for i in WED:
    WED_WT.append(i.text)
```

## 목요일 웹툰 

```python
THU = scrap_html(url).find('div', {
    'class': 'daily_section _list_THURSDAY'
}).find_all('p', "subj")
print(len(THU))
THU_WT = []
for i in THU:
    THU_WT.append(i.text)
```

## 금요일 웹툰 

```python
FRI = scrap_html(url).find('div', {
    'class': 'daily_section _list_FRIDAY'
}).find_all('p', "subj")
print(len(FRI))
FRI_WT = []
for i in FRI:
    FRI_WT.append(i.text)
```

## 토요일 웹툰 

```python
SAT = scrap_html(url).find('div', {
    'class': 'daily_section _list_SATURDAY'
}).find_all('p', "subj")
print(len(SAT))
SAT_WT = []
for i in SAT:
    SAT_WT.append(i.text)
```

## 일요일 웹툰 

```python
SUN = scrap_html(url).find('div', {
    'class': 'daily_section _list_SUNDAY'
}).find_all('p', "subj")
print(len(SUN))
SUN_WT = []
for i in SUN:
    SUN_WT.append(i.text)
```

# 요일별 웹툰 인기순 DataFrame

```python
days = [MON_WT, TUE_WT, WED_WT, THU_WT, FRI_WT, SAT_WT, SUN_WT]
df = pd.DataFrame(days,
                  index=['MON', 'TUE', 'WED', 'THU', 'FRI', 'SAT', 'SUN'])
df = df.transpose()
df
```

## 전체 웹툰 수, 중복 제거 수 

```python
count = 0
for i in days:
    count += len(i)

dou = []
for i in days:
    for j in i:
        dou.append(j)

print('전체 :', count, ' \n중복제거 : ', len(list(set(dou))))
```

# 웹툰 등급화 분류(1~4등급) 

```python
Class1, Class2, Class3, Class4 = [], [], [], []
for day in days:
    for WT in day[:9]:
        Class1.append(WT)

for day in days:
    for WT in day[9:18]:
        Class2.append(WT)

for day in days:
    for WT in day[18:27]:
        Class3.append(WT)

for day in days:
    for WT in day[27:]:
        Class4.append(WT)
```

```python
# 각 등급 개수
print(len(Class1), len(Class2), len(Class3), len(Class4))
```

```python
# 각 등급 중복 제거 개수
print(len(list(set(Class1))), len(list(set(Class2))), len(list(set(Class3))),
      len(list(set(Class4))))
```

# 등급 분류 웹툰 DataFrame

```python
Class = [
    list(set(Class1)),
    list(set(Class2)),
    list(set(Class3)),
    list(set(Class4))
]
df_class = pd.DataFrame(Class, index=['1', '2', '3', '4']).transpose()
df_class = df_class.fillna('-')
df_class
```

# 베스트댓글 좋아요 수 feature 스케일링

```python
df_EN = pd.read_csv('[영어]인기순.csv', encoding='utf-8')
df_EN
```

```python
comments = pd.DataFrame(df_EN['Comments Likes'].values, columns=['좋아요 수'])

from sklearn.preprocessing import StandardScaler
s_scaler = StandardScaler()
comments = s_scaler.fit_transform(comments)

df_EN.loc[:, 'Comments Likes'] = comments
```

# 웹툰 등급 라벨링(Target)

```python
no = []
```

```python
for i in list(df_class['1']):
    if len(df_EN[df_EN['Title'] == i]) > 0:
        ind = df_EN[df_EN['Title'] == i].index[0]
        df_EN.loc[ind, 'Target'] = 1
    elif i == '-':
        pass
    else:
        print('없는 작품 :', i)
        no.append(i)
```

```python
for i in list(df_class['2']):
    if len(df_EN[df_EN['Title'] == i]) > 0:
        ind = df_EN[df_EN['Title'] == i].index[0]
        df_EN.loc[ind, 'Target'] = 2
    elif i == '-':
        pass
    else:
        print('없는 작품 :', i)
        no.append(i)
```

```python
for i in list(df_class['3']):
    if len(df_EN[df_EN['Title'] == i]) > 0:
        ind = df_EN[df_EN['Title'] == i].index[0]
        df_EN.loc[ind, 'Target'] = 3
    elif i == '-':
        pass
    else:
        print('없는 작품 :', i)
        no.append(i)
```

```python
for i in list(df_class['4']):
    if len(df_EN[df_EN['Title'] == i]) > 0:
        ind = df_EN[df_EN['Title'] == i].index[0]
        df_EN.loc[ind, 'Target'] = 4
    elif i == '-':
        pass
    else:
        print('없는 작품 :', i)
        no.append(i)
```

```python
# 해당 데이터내에 없는 웹툰 확인
print(len(no),'\n',no)
```
