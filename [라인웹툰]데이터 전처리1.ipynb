# 좋아요 수 합치기 
웹툰에 대한 좋아요 수는 웹툰 상세페이지에서 얻을 수 없다.  
따라서 기존에 구했던 데이터 파일을 읽어와 각 웹툰(title)에 맞는 좋아요 수를 새로운 컬럼(Total Likes)에 추가한다.
* df1 : 좋아요 수 가져올 데이터 
*  df2 : 좋아요 수 넣을 데이터

```python
import pandas
```

```python
df1 = pd.read_csv('웹툰_Likes_0520.csv',encoding='utf-8')  # 좋아요 수 가져올 데이터 
df2 = pd.read_csv('./그림체분류/[영어]드라마장르_Drama_그림체.csv',encoding='utf-8') #좋아요 수 넣을 데이터

for i in df2['Title']:
    
    if len(df1[df1['Title'].str.contains(i)]) >0 :# df2의 웹툰이 df1에 있을 경우
        Likes = df1[df1['Title']==i]['Likes'].values[0] # 해당 웹툰의 좋아요 수 
        row_ = df2[df2['Title']==i].index[0]        # df2의 index를 찾는다.  
        df2.loc[row_,'Total Likes'] = Likes   # df2의 total Likes에 저장한다.
    else:
        pass
print(df2[df2['Total Likes'].isna()])
df_Drama = df2
df_Drama['Genre'] = 'Drama'
df_Drama.to_csv('./data/[영어]드라마장르_Drama_그림체.csv',encoding='utf-8',index=False)
```
